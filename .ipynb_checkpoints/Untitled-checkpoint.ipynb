```python
import lightkurve as lk
import numpy as np
import matplotlib.pyplot as plt
```

```python
p, t0 = 1.7554137, 1573.111260000
def double_up(lc):
    f = lc.fold(p, t0)
    f2 = lc.fold(p, t0)
    f2.time += 1
    return f.append(f2)
```

```python
tpf = lk.search_targetpixelfile('Wasp-104').download()
```

```python
f = tpf.to_lightcurve().fold(p/2, t0)
mask = np.in1d(tpf.to_lightcurve().time, f.time_original[np.abs(f.phase) < 0.1])
_, m = tpf.to_lightcurve()[~mask].flatten(101).remove_outliers(return_mask=True)
mask[np.in1d(tpf.time, tpf.time[~mask][m])] = False
```

```python
def elipsoidal(t, A):
    return -A * np.cos
```

```python
phase = (tpf.time - t0 - p/2) / p % 1 - 0.5
phase2 = np.copy(phase) + 0.5
phase2[phase > 0] -= 1
plt.plot(tpf.time, np.cos(4 * np.pi * phase)*0.01 + 1)
plt.plot(tpf.time, np.cos(2 * np.pi * phase2)*0.01 + 1)


plt.plot(tpf.time, tpf.to_lightcurve().normalize().flux)
plt.xlim(tpf.time[0], tpf.time[200])

add = np.vstack([np.cos(4 * np.pi * phase), np.cos(2 * np.pi * phase2)]).T
#add = np.atleast_2d(np.cos(2 * np.pi * phase2)).T
```

```python
c = tpf.to_corrector(aperture_mask=tpf.pipeline_mask, pld_aperture_mask='all', pld_order=2, n_pca_terms=25, additional_columns=add)
```

```python
lc = c.correct(remove_gp_trend=False, cadence_mask=~mask, pld_order=2, npca_terms=25, gp_timescale_prior=300, additional_columns=add)
```

```python
n = 1
motion = np.reshape(np.dot(c.design_matrix[:, :-n], c.most_recent_solution_or_trace['weights'][:-n]), -1)
gp_flux = c.get_diagnostic_lightcurves(c.most_recent_solution_or_trace)[1].flux
```

```python
phase = lk.LightCurve(tpf.time, np.reshape(np.dot(c.design_matrix[:, -n:], c.most_recent_solution_or_trace['weights'][-n:]), -1))
```

```python
lc = tpf.to_lightcurve(aperture_mask=tpf.pipeline_mask)
clc = (lc - gp_flux - motion)
```

```python
ax = double_up(clc).bin(15).errorbar(normalize=False)
double_up(phase).plot(normalize=False, ax=ax, lw=2, label='Best Fit Phase Curve')
plt.ylim(-30, 30)
```
