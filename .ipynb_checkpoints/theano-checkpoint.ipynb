```python
import theano.tensor as tt
import numpy as np
import matplotlib.pyplot as plt
import pymc3 as pm
import exoplanet as xo
import starry
```

```python
import astropy.units as u
from astropy.constants import G
from astropy.modeling.blackbody import blackbody_lambda
```

```python
bandpass = np.loadtxt('data/bandpass.dat')
bandpass[:, 1] /= np.trapz(bandpass[:,1], bandpass[:,0])
```

```python
time = np.linspace(0, 10, 10000)

# Star Guesses
m_star = 0.5
r_star = 1
teff = 3400
norm = np.sum(blackbody_lambda(bandpass[:, 0]*u.nm, teff*u.K).value * bandpass[:, 1])


```

```python
def elipsoidal(t, P, t0, A):
    return -A * np.cos((2 * np.pi * (t-t0))/(0.5 * P))

def doppler(t, P, t0, A):
    return A * np.sin((2 * np.pi * (t-t0))/(P))

def reflection(t, albedo, radius, r_star, a, period, t0):
    A = albedo * (((radius*r_star*u.solRad)**2)/((a*u.AU)**2).to(u.solRad**2)).value 
    return -A * np.cos((2 * np.pi * (t-t0))/(period))

def equilibrium_temp(r_star, a, teff):
    a1 = (((r_star*u.solRad))/((a*u.AU)).to(u.solRad)).value
    return teff * a1**0.5 * (0.25 * (1-(albedo*2/3)))**(1/4)


def flux_in_bandpass(albedo, r_star, a, teff):
    teq = equilibrium_temp(r_star, a, teff)
    f_in_bp = np.sum(blackbody_lambda(bandpass[:, 0]*u.nm, teq*u.K).value * bandpass[:, 1])
    f_in_bp /= norm
    return f_in_bp
    
def thermal(t, albedo, r_star, a, teff, period, t0, phase_shift):
    f_in_bp = flux_in_bandpass(albedo, r_star, a, teff)
    return -f_in_bp * np.cos((2 * np.pi * (t-t0))/(period) + phase_shift)
```

```python
plt.plot(time, thermal(time, 1, 1, 0.00838, 3479, 0.448, 0, 0) + 1)
#plt.plot(time, refle(time, 1, 1, 0.00838, 3479, 0.448, 0, 0) + 1)
```

```python
# Starting Guesses
# RADIUS IS OVER R_STAR
radius_value = 0.1
period_value = 0.448
t0_value = 0
albedo_value=0.1
#flux_ratio_value = 1e-4

a = ((((period_value * u.day)**2 * G * (m_star*u.solMass)/(4*np.pi**2))**(1/3))).to(u.AU).value
#eccentricity = 0
#omega = 0.5*np.pi
inclination = 90*(np.pi/180)

el = 0.00001
do = 0.00001
re =  albedo_value * (((radius_value*r_star*u.solRad)**2)/((a*u.AU)**2).to(u.solRad**2)).value
phase_shift = 0
u_value = [0.2, 0.3]
f_in_bp = flux_in_bandpass(albedo_value, r_star, a, teff)
teq = equilibrium_temp(r_star, a, teff)

eclipse_t0 = t0_value + period_value/2
logP = np.log(period_value)

```

**assume no eccentricity**

```python
transit_orbit = xo.orbits.KeplerianOrbit(period=period_value,
                                         t0=t0_value,
                                         incl=inclination,
                                            a=a*u.AU.to(u.solRad)/r_star)
transit = xo.StarryLightCurve(u_value).get_light_curve(
                            orbit=transit_orbit, r=radius_value, t=time)
transit = pm.math.sum(transit, axis=-1)
```

```python
eclipse_orbit = xo.orbits.KeplerianOrbit(period=period_value, 
                                         t0=eclipse_t0,
                                         incl=inclination,
                                         a=a*u.AU.to(u.solRad)/r_star)
eclipse = xo.StarryLightCurve([0, 0]).get_light_curve(
                            orbit=eclipse_orbit, r=radius_value, t=time)
eclipse = pm.math.sum(eclipse, axis=-1)
eclipse = ((eclipse)/radius_value**2) * (2 * re + f_in_bp)
```

```python
data = (elipsoidal(time, period_value, t0_value, el) +
        doppler(time, period_value, t0_value, do) +
        reflection(time, albedo_value, radius_value, r_star, a, period_value, t0_value) + 
        thermal(time, albedo_value, r_star, a, teff, period_value, t0_value, 0))
data += eclipse.eval()
data += transit.eval()
data += 1
data += np.random.normal(0, 0.00001, len(time))
#data += sys.lightcurve
err = np.zeros(len(time)) + 0.00001
```

```python
plt.errorbar(time, data, err, ls='')
plt.xlim(0, 1)
```

```python
def build_model(mask=None, start=None):
    
    if mask is None:
        mask = np.ones(len(time), dtype=bool)
        
    with pm.Model() as model:
        
        
        M_star = pm.Normal("M_star", mu=m_star, sd=0.01)
        R_star = pm.Normal("R_star", mu=r_star, sd=0.01)
        ## SHOULD ADD TEFF HERE BUT NEED THEANO IMPLEMENTATION OF BB 
        
        mean = pm.Normal("mean", mu=1.0, sd=10.0)
        logP = pm.Normal("logP", mu=np.log(period_value), sd=0.5)
        period = pm.Deterministic("period", pm.math.exp(logP))
        t0 = pm.Normal("t0", mu=t0_value, sd=1.0)
        u = xo.distributions.QuadLimbDark("u", testval=np.array(u_value))
        
        logr = pm.Normal("logr", mu=np.log(radius_value), sd=1.0)
        pm.Potential("r_prior", -logr)
        r = pm.Deterministic("r", tt.exp(logr))

        
        
        
        # Transit
        transit_orbit = xo.orbits.KeplerianOrbit(period=period, t0=t0, b=0)
        transit = xo.StarryLightCurve([0.2, 0.3]).get_light_curve(
                                    orbit=transit_orbit, r=r, t=time)
        transit = pm.math.sum(transit, axis=-1)

        # Phase Curve
        el = pm.Normal("el", mu=0, sd=1)
        do = pm.Normal("do", mu=0, sd=1)
        re = pm.Normal("re", mu=0, sd=1)
        re_ph = pm.Normal("re_ph", mu=0, sd=1)
    
        elipsoidal = pm.Potential('elipsoidal', -el * pm.math.cos((2 * np.pi * (time-t0))/(0.5 * tt.exp(logP))))
        doppler = pm.Potential('doppler', do * pm.math.sin((2 * np.pi * (time-t0))/(tt.exp(logP))))
        reflection = pm.Potential('reflection', -re * pm.math.cos((2 * np.pi * (time-t0))/(tt.exp(logP)) + re_ph))

        
        # Secondary Eclipse
        eclipse_orbit = xo.orbits.KeplerianOrbit(period=period, t0=t0 + period/2, b=0)
        eclipse = xo.StarryLightCurve([0, 0]).get_light_curve(
                                    orbit=eclipse_orbit, r=r, t=time)
        eclipse = pm.math.sum(eclipse, axis=-1)
        eclipse = ((eclipse)/r**2) * 2 * re
        pm.Deterministic("eclipse", eclipse)
    
    
        # Build the light curve
        light_curve = elipsoidal + doppler + reflection + mean + eclipse + transit
        pm.Deterministic("light_curve", light_curve)
        pm.Normal("obs", mu=light_curve, sd=err, observed=data)

        # Optimize
        if start is None:
            start = model.test_point
        map_soln = xo.optimize(start=start, vars=[mean])
        map_soln = xo.optimize(start=map_soln, vars=[logP, t0, r])
        map_soln = xo.optimize(start=map_soln, vars=[re, logP, t0, r])
        map_soln = xo.optimize(start=map_soln, vars=[re, re_ph, logP, mean])
        map_soln = xo.optimize(start=map_soln, vars=[re, re_ph, do, el, logP, mean])
        map_soln = xo.optimize(start=map_soln, vars=[u, r])
        
        map_soln = xo.optimize(start=map_soln)




    return model, map_soln
```

```python
model, map_soln = build_model()
```

```python
plt.plot(time, data)
with model:
    print('P:', np.exp(map_soln['logP']))
    plt.plot(time, map_soln['light_curve'])
```

```python
with model:
    plt.errorbar(time, data - map_soln['mean'], err, ls='', c='k', alpha=0.2)
    plt.plot(time, elipsoidal(time, np.exp(map_soln['logP']), map_soln['el']))
    plt.plot(time, doppler(time, np.exp(map_soln['logP']), map_soln['do']))
    plt.plot(time, reflection(time, np.exp(map_soln['logP']), map_soln['re'], map_soln['re_ph']) + map_soln['eclipse'])
```
